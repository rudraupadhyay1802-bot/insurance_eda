```python
# Creating a count plot to visualize the distribution of 'sex'
sns.countplot(x="sex", data=df)
plt.title("Gender Distribution")
plt.show()
```



*  Dataset has a nearly equal distribution of males and females
*  No gender imbalance present



```python
# Creating a count plot to visualize the distribution of 'region'
sns.countplot(x='region', data=df)
plt.title("Region Distribution")
plt.show()
```



*   All regions are almost equally represented
*   No region dominates the dataset



```python
# Creating a scatter plot to visualize the relationship between 'age' and 'charges'
sns.scatterplot(x="age", y="charges", data=df)
plt.title("Age vs Charges")
plt.show()
```



*   Positive relationship observed
*   As age increases, medical charges tend to increase
*   Older people generally have higher charges





```python
# Creating a scatter plot to visualize the relationship between 'bmi' and 'charges'
sns.scatterplot(x="bmi", y="charges", data=df)
plt.title("BMI vs Charges")
plt.show()
```



*   Slight positive trend between BMI and charges
*   Higher BMI may lead to increased charges
*   Relationship is not very strong





```python
# Creating a heatmap to visualize the correlation matrix of numerical features
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()
```

*   Age and charges show moderate correlation
*   BMI has weak to moderate correlation with charges
*   Other variables show low correlation





```python
plt.hist(df["charges"], bins=10)
plt.xlabel("Charges")
plt.ylabel("Count")
plt.title("Histogram of Charges")
plt.show()
```

*   The distribution is right-skewed
*   Most individuals have low to moderate charges
*   A few cases show very high charges (outliers)







```python
# Creating a bar chart to visualize the distribution of 'smoker' status
df["smoker"].value_counts().plot(kind="bar")
plt.xlabel("Smoker")
plt.ylabel("Count")
plt.title("Bar Chart of Smoker")
plt.show()
```



*   Majority of individuals are non-smokers
*   Smokers represent a smaller portion of the dataset



```python
# Creating a box plot to visualize the distribution of 'charges' based on 'smoker' status
sns.boxplot(x="smoker", y="charges", data=df)
plt.title("Box Plot of Charges by Smoker")
plt.show()
```



*   Smokers have significantly higher charges than non-smokers
*   Strong indication that smoking impacts charges



```python
# Creating a box plot to visualize the distribution of 'charges' based on 'region'
sns.boxplot(x="region", y="charges", data=df)
plt.title("Box Plot of Charges by Region")
plt.show()
```

*   Slight variation across regions
*   No region shows consistently extreme charges
*   Region has minimal impact on cost





```python
# Creating a box plot to visualize the distribution of 'charges' based on 'sex'
sns.boxplot(x="sex", y="charges", data=df)
plt.title("Box Plot of Charges by Gender")
plt.show()
```



* Charges are similar across genders
* Gender does not significantly affect charges



```python
# Creating a pair plot to visualize relationships between all numerical variables and distributions
sns.pairplot(df)
plt.show()
```

* Clear separation seen for smoker vs charges

* Confirms relationships between variables

* Helps visualize overall data patterns and trends



These plots help to understand data distribution, relationships, and key patterns before applying machine learning.
